```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("admission.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot: GRE Score vs. Chance of Admit
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="GRE Score", y="Chance of Admit", hue="Research", palette="viridis")
plt.title("GRE Score vs. Chance of Admit", fontsize=14)
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")
plt.legend(title="Research")
plt.show()

```

```python
# Bar Plot: Average CGPA by University Rating
plt.figure(figsize=(8, 6))
avg_cgpa = df.groupby("University Rating")["CGPA"].mean().reset_index()
sns.barplot(data=avg_cgpa, x="University Rating", y="CGPA", palette="magma")
plt.title("Average CGPA by University Rating", fontsize=14)
plt.xlabel("University Rating")
plt.ylabel("Average CGPA")
plt.show()
```

```python
# Box Plot: Distribution of SOP Scores by Research
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Research", y="SOP", palette="coolwarm")
plt.title("Distribution of SOP Scores by Research", fontsize=14)
plt.xlabel("Research (0=No, 1=Yes)")
plt.ylabel("SOP Score")
plt.show()
```

```python
# Pie Chart: Percentage of Applicants with and without Research
plt.figure(figsize=(8, 8))
research_counts = df["Research"].value_counts()
labels = ["No Research", "Has Research"]
plt.pie(research_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=["lightcoral", "skyblue"])
plt.title("Percentage of Applicants with and without Research", fontsize=14)
plt.show()
```

```python
# Line Chart: Average Chance of Admit by GRE Score
plt.figure(figsize=(8, 6))
avg_admit = df.groupby("GRE Score")["Chance of Admit"].mean().reset_index()
sns.lineplot(data=avg_admit, x="GRE Score", y="Chance of Admit", color="darkgreen")
plt.title("Average Chance of Admit by GRE Score", fontsize=14)
plt.xlabel("GRE Score")
plt.ylabel("Average Chance of Admit")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["CGPA"], bins=20, kde=True, color="royalblue")
plt.title("Distribution of CGPA Scores", fontsize=14)
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x="University Rating", y="GRE Score", palette="pastel")
plt.title("GRE Score Distribution by University Rating", fontsize=14)
plt.xlabel("University Rating")
plt.ylabel("GRE Score")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="University Rating", palette="viridis")
plt.title("Number of Applicants per University Rating", fontsize=14)
plt.xlabel("University Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Research", y="CGPA", palette="coolwarm")
plt.title("CGPA Distribution by Research Experience", fontsize=14)
plt.xlabel("Research (0=No, 1=Yes)")
plt.ylabel("CGPA")
plt.show()
```
